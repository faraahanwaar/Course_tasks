# Yalla Kora

```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.yallakora.com/match-center/%D9%85%D8%B1%D9%83%D8%B2-%D8%A7%D9%84%D9%85%D8%A8%D8%A7%D8%B1%D9%8A%D8%A7%D8%AA?date=8/25/2023#"
```

```python
v=requests.get(url)
content=v.content
content
```

```python
soup=BeautifulSoup(content,"html.parser")
soup
```

```python
matches = soup.find_all('div', class_='matchCard')

len(matches)
```

```python
for i in range(len(matches)):
    print(matches[i].find("h2").text.strip())
    teams=matches[i].find_all("li")
    for j in range(len(teams)):
        teamA=teams[j].find("div" ,class_="teams teamA").text.strip()
        teamB=teams[j].find("div" ,class_="teams teamB").text.strip()
        
        results = teams[j].find_all('span', class_='score')
        score = f'{results[0].text.strip()} - {results[1].text.strip()}'
        
        time=teams[j].find("span", class_="time").text.strip()
        matchStatus=teams[j].find("div", class_="matchStatus").text.strip()
        
        
        if teams[j].find("div", class_="channel icon-channel"):
            channel=teams[j].find("div", class_="channel icon-channel").text.strip()
        else:
            channel=None
            
        date=teams[j].find("div" ,class_="date").text.strip()
         
        print(f"\n {date} :date  ") 
        print(teamA,score,teamB,time,channel)
        print(f" {matchStatus} :matchStatus \n ")                        
    
    print("="*20)
        
        
```
